## DATA CLEANING

- BookingID column is same of index table, we will delete it
- Change target column from categorical to numerical

```python
df_c = df.copy()
```

```python
df_c = df_c.drop(columns='Booking_ID', axis=1)
df_c['booking_status'] = df_c['booking_status'].replace(['Not_Canceled', 'Canceled'], [0,1])
df_c.head()
```

## EDA

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
temp = ['lead_time', 'avg_price_per_room', 'arrival_date', 'no_of_previous_bookings_not_canceled']

for column_name in df_c.drop(columns=temp).columns:
    sns.countplot(data=df_c, x=column_name)
    plt.title(column_name)
    plt.show()
```

## FEATURE TRANSFORMATION

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.model_selection import train_test_split
```

```python
x = df_c.drop(columns="booking_status", axis=1)
y = df_c['booking_status']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=20)
```

```python
step1 = ColumnTransformer([
    ("step1", OneHotEncoder(), ['type_of_meal_plan', 'room_type_reserved', 'market_segment_type'])
], remainder='passthrough')
```

## ML MODELS

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

```python
models = [
    LogisticRegression(),
    RandomForestClassifier(),
    XGBClassifier(n_estimators=100, max_depth=6, learning_rate=0.1, objective='binary:logistic')
]
```

```python
for model in models:
    pipe = Pipeline([
        ("step1", step1),
        ("model", model)
    ])

    pipe.fit(x_train, y_train)
    predict = pipe.predict(x_test)

    print(f"Acc: {round(accuracy_score(y_test, predict)*100,2)} %")
    print(classification_report(y_test, predict))
    print(confusion_matrix(y_test, predict))
```

- RF model has better performance in contrast to others (90% acc aprox) and acceptable f1 score
- RF model has been chosen

```python
model = RandomForestClassifier()
pipe = Pipeline([
        ("step1", step1),
        ("model", model)
    ])
pipe.fit(x_train, y_train)
predict = pipe.predict(x_test)

print(f"Acc: {round(accuracy_score(y_test, predict)*100,2)} %")
```

## FEATURE SELECTION

```python
from sklearn.inspection import permutation_importance
r = permutation_importance(pipe, x_test, y_test,
                           n_repeats=10,
                           random_state=0)
perm = pd.DataFrame(columns=['AVG_Importance', 'STD_Importance'], index=[i for i in x_train.columns])
perm['AVG_Importance'] = r.importances_mean
```

```python
sns.barplot(data=perm, y=perm.index, x='AVG_Importance', order=perm.sort_values("AVG_Importance", ascending=False).index)
plt.show()
```

- Many features do not have a major impact on the ml model
- These features will be removed
- Then, it is expected that the ML model presents the same metrics or that it does not get worse

## RETRAIN WITH LESS FEATURES

```python
# DELETE THIS FEATURES

df_new_feat = df_c.drop(columns=["no_of_children", "repeated_guest", "no_of_previous_cancellations","required_car_parking_space","no_of_previous_bookings_not_canceled"], axis=1)
```

```python
x_new_feat = df_new_feat.drop(columns="booking_status", axis=1)
y_new_feat = df_new_feat['booking_status']
x_train, x_test, y_train, y_test = train_test_split(x_new_feat,y_new_feat,test_size=0.2,random_state=20)
```

```python
for model in models:
    pipe = Pipeline([
        ("step1", step1),
        ("model", model)
    ])

    pipe.fit(x_train, y_train)
    predict = pipe.predict(x_test)

    print(f"Acc: {round(accuracy_score(y_test, predict)*100,2)} %")
    print(classification_report(y_test, predict))
    print(confusion_matrix(y_test, predict))
```

- Again, Random Forest model shows best performance and with less features
- Also, RF model present almost same metrics

```python
model = RandomForestClassifier()
pipe = Pipeline([
        ("step1", step1),
        ("model", model)
    ])
pipe.fit(x_train, y_train)
predict = pipe.predict(x_test)

print(f"Acc: {round(accuracy_score(y_test, predict)*100,2)} %")
```
